# **OVERVIEW**

With Android powering approximately 85% of the global smartphone market, the Google Play Store stands as the leading distribution platform for Android apps. This project aims to explore and interpret detailed metadata about Play Store apps to uncover trends, user behaviors, and app performance factors. By leveraging exploratory data analysis (EDA), the goal is to generate insights that can support developers and businesses in refining their app strategies and enhancing user engagement.

The study not only examines app-level attributes such as category, size, pricing, and ratings, but also integrates user-generated review data to investigate the emotional tone and sentiment of customer feedback.

# **About Dataset**

Contains information on over 10,000 Android applications, including details such as app name, category, rating, number of installs, price, content rating, size, and type (Free or Paid). The dataset consists of 13 structured columns.

# **Project Summary**
This project analyzes Google Play Store app data to uncover trends in app performance, user engagement, and market behavior. Using Python libraries like Pandas, Matplotlib, and Seaborn, we explored relationships between app features such as category, installs, rating, price, and type. The goal is to identify what makes an app successful in a highly competitive Android ecosystem.

# **Problem Statement**
To examine Play Store data and discover key attributes that influence app popularity and user engagement.

# **Business Objective**
Provide data-driven insights to help developers understand user behavior, optimize app features, and align their strategies for higher installs and better app visibility.

### **Data Cleaning**

Handling Null Values

```python
df.isnull().sum()
```

Observation - As the Scatter Plot of Rating column is right skewed, will use Median as our immutation technique.

```python
col = df['Rating']
col.fillna(col.median(), inplace = True)
```

```python
col.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Type'], inplace=True)
df.dropna(subset=['Content Rating'], inplace=True)
df.dropna(subset=['Current Ver'], inplace=True)
df.dropna(subset=['Android Ver'], inplace=True)
```

```python
df.isnull().sum()
```

Handling Duplicate Value

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

Cleaning Installs column

```python
df['Installs'].value_counts()
```

```python
df['Installs'] = df['Installs'].str.replace("+","").str.replace(",", "").astype(int)
```

```python
df['Installs'].value_counts()
```

```python
np.dtype(df['Installs'])
```

```python
df['Installs'].unique()
```

Changing type of revies column

```python
print(np.dtype(df['Reviews']))
```

```python
df['Reviews'] = df['Reviews'].astype(int)
print(np.dtype(df['Reviews']))
```

Cleaning Size Column

```python
df['Size'].value_counts()
```

```python
df['Size'].unique()
```

Converting all sizes into MB

```python
def fix_size(s):
  if isinstance(s, str):
    if 'k' in s:
      return float(s.replace("k", "")) / 1024
    elif 'M' in s:
      return float(s.replace("M", ""))
    elif 'Varies with device' in s:
      return np.nan
  return s
```

```python
df['Size'] = df['Size'].apply(fix_size)
df['Size'].unique()
```

Formating Price

```python
df['Price'].value_counts()
```

```python
def new_price(price):
  if isinstance(price, str):
    if "$" in price:
      return price.replace("$", "")
    return price
df['Price'] = df['Price'].apply(new_price)
```

```python
df['Price'].value_counts()
```

```python
df['Price'].dtype
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df['Price'].dtype
```
